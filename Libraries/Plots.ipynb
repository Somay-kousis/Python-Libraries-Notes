```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2) 

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
#plot t vs t, t vs t2, t vs t3
plt.show()
```

```python
data = {
    'a': np.arange(50),  # 0 to 49 (X-axis values)
    'c': np.random.randint(0, 50, 50),  # random integers (used for color)
    'd': np.random.randn(50)  # 50 random normal values (used for size)
}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)

plt.subplot(132)
plt.scatter(names, values)

plt.subplot(133)
plt.plot(names, values)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
# 'bo' → blue circle markers for t1
# 'k' → black solid line for t2
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
mu, sigma = 100, 15  # Mean and standard deviation
x = mu + sigma * np.random.randn(10000)  # Generate 10,000 random values (normal distribution)

# Plot histogram
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')             # X-axis label
plt.ylabel('Probability')        # Y-axis label
plt.title('Histogram of IQ')     # Plot title
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')  # Add mean and std text
plt.axis([40, 160, 0, 0.03])     # Set axis limits
plt.grid(True)                   # Show grid
plt.show()                       # Display plot
```

| Part            | Meaning                                                                |
| --------------- | ---------------------------------------------------------------------- |
| `x`             | The data you're plotting (IQ values here)                              |
| `50`            | Number of bins (divides data into 50 bars)                             |
| `density=True`  | Normalizes the histogram (area under it = 1, like a probability curve) |
| `facecolor='g'` | Color of the bars (green)                                              |
| `alpha=0.75`    | Transparency of bars (1 = solid, 0 = invisible)                        |

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```
